```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Triest base vs sample size

```python
M = [3000, 5000, 7000, 9000, 10000, 12000]
T = [2015982, 1554563, 1404758, 1471005, 1297671, 1353472]

plt.plot(M,T,'bo',linestyle='dashed')
plt.axhline(y=1351441, linewidth=2, color='g')
plt.xlabel('Sample Size')
plt.ylabel('Triangles estimated')
plt.title('TRIEST BASE: Triangles estimated Vs Sample Size')
plt.ylim((1200000,3000000))
```
